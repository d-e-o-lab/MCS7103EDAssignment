```python
X = new_data.drop('Target', axis=1)
y = new_data['Target']
```

```python

# Splitting into train and test datasets
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Create an  Decision Tree classifier
dectree = DecisionTreeClassifier(random_state=0)
# Create an Random Forest classifier
ranfor = RandomForestClassifier(random_state=2)
# Create an Logistic Regression classifier
logreg = LogisticRegression(random_state=42)
# Create an K-Nearest Neighbors classifier
knn = KNeighborsClassifier(n_neighbors=3)
# Create an SVM classifier
svm = svm.SVC(kernel='linear',probability=True)
```

```python
# Train the decision tree classifier
dectree.fit(X_train,y_train)
# Train the random forest classifier
ranfor.fit(X_train,y_train)
# Train the logistic regression classifier
logreg.fit(X_train,y_train)
# Train the K-Nearest Neighbors classifier
knn.fit(X_train,y_train)
# Train the Support vector Machine classifier
svm.fit(X_train, y_train)
```

```python
# Calculate Decision Tree accuracy
y_pred = dectree.predict(X_test)
# Calculate Decision Tree accuracy
print("Accuracy :",round(accuracy_score(y_test,y_pred)*100,2),"%")
precision = precision_score(y_test, y_pred,average='macro')
print("Precision :",round(precision*100,2),"%")
recall = recall_score(y_test, y_pred,average='macro')
print("Recall :",round(recall*100,2),"%")
f1 = f1_score(y_test, y_pred,average='macro')
print("F1 Score :",round(f1*100,2),"%")
```
